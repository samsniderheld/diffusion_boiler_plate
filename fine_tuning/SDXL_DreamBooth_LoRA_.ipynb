```python
import glob
import json
dir = "/content/Black_Council_Dragon_Rider_Female_Fixed"
paths = sorted([path for path in glob.glob(f"{dir}/*.jpg")])

caption_file = "/content/HOTD_Experiments/fine_tuning/dragon_rider_captions.txt"

with open(f'{dir}/metadata.jsonl', 'w') as outfile:
  with open(caption_file, 'r') as f:
    lines = f.readlines()
    print(len(lines),len(paths))
    for i, line in enumerate(lines):
      entry = {"file_name":paths[i].split("/")[-1], "prompt": line}
      json.dump(entry, outfile)
      outfile.write('\n')
```

```python
base_path = "SG161222/RealVisXL_V4.0"
additional_controlnet_paths = [
    "thibaud/controlnet-openpose-sdxl-1.0",
    "diffusers/controlnet-depth-sdxl-1.0",
]

additional_loras = [
    {
      "model_id":"/content/Black_Council_Dragon_Rider_Female_2000",
      "filename":"pytorch_lora_weights.safetensors"
     }
]

pipe = HOTD_SDXL_Pipeline(base_path, additional_controlnet_paths,additional_loras, use_refiner=False, use_hires_fix=False)
pipe.load_pipeline()
```
